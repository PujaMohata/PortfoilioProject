```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv(r'C:\Users\dell\Desktop\Internship\movies.csv')
```

```python
movie.head()
```

```python
movie['genres'] = movie.genres.str.split('|')
movie.head()
```

Now that genres is in a list format, will use this list to convert into a vector

```python
#Copying the dataframe into a new one since genre info. is not needed in future implementation
movies_copy = movie.copy()
```

```python
#now iterating thru the list of genres and placing 1 into corresponding movies list
for index, row in movie.iterrows():
    for genre in row['genres']:
        movies_copy.at[index,genre] = 1
```

```python
#Filling nan values with 0
movies_copy = movies_copy.fillna(0)
movies_copy.head()
```

```python
ratings = pd.read_csv(r'C:\Users\dell\Desktop\Internship\ratings.csv')
```

```python
ratings.head()
```

```python
# taking user input to recommend movies to
user_input = [{'title':'Toy Story (1995)','rating':5},
             {'title':'Jumanji (1995)','rating':4},
             {'title':'Pulp Fiction (1994)','rating':3.5},
             {'title':'Akira (1988)','rating':4.5},
             {'title':'Breakfast Club, The (1985)','rating':2}]
inputmovie= pd.DataFrame(user_input)
inputmovie
```

```python
# add movieid to input user
#filtering movies by title
inputid=movie[movie['title'].isin(inputmovie['title'].tolist())]
```

```python
#then merging it to get movie id
inputmovie = pd.merge(inputid,inputmovie)
inputmovie
```

```python
#dropping genre from the dataframe
#inputmovie = inputmovie.drop('genres',1)
inputmovie
```

Getting genre table for user movies

```python
#filtering out the movies from the input
userMovie = movies_copy[movies_copy['movieId'].isin(inputmovie['movieId'].tolist())]
userMovie
```

```python
#Resetting the index
userMovie = userMovie.reset_index(drop=True)

#dropping all the columns except the diff genres
userGenreTable = userMovie.drop('movieId',1).drop('title',1).drop('genres',1)
userGenreTable
```

From the input prefernce we will see the inference, by turning each genre into weights which will be done by mutliplying input ratings to input genre table

```python
inputmovie['rating']
```

```python
#dot product to get weights
userProfile = userGenreTable.transpose().dot(inputmovie['rating'])
userProfile
```

As the weights are there for every of the user's preference now we can recommend movies accordingly

```python
#getting genres of every movie in original dataframe
genre_table = movies_copy.set_index(movies_copy['movieId'])

#dropping all othe columns
genre_table = genre_table.drop('movieId',1).drop('title',1).drop('genres',1)
genre_table.head()
```

```python
genre_table.shape
```

```python
#Multiplying the genre by the weights and then taking mean
recommender = ((genre_table*userProfile).sum(axis=1))/(userProfile.sum())
recommender.head()
```

```python
#Sort recommendation
recommender = recommender.sort_values(ascending=False)
recommender.head()
```

```python
#the Final Recommendation
movie.loc[movie['movieId'].isin(recommender.head(20).keys())]
```
