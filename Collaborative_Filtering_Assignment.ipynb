```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = pd.read_csv(r'C:\Users\dell\Desktop\Internship\movies_metadata.csv',low_memory=False)
```

```python
movie_df = movies.rename(columns={"id": "movieId"})
```

```python
movie_df.head()
```

```python
ratings_data = pd.read_csv(r'C:\Users\dell\Desktop\Internship\ratings_small.csv')
ratings_data.head()
```

```python
ratings_data.dtypes
```

```python
ratings_data['movieId'] = ratings_data.movieId.astype(str)
```

```python
ratings_data.dtypes
```

```python
movie_data = pd.merge(movie_df,ratings_data, on='movieId')
```

```python
movie_data.head()
```

```python
#creating mean ratings data
rating = pd.DataFrame(movie_data.groupby('title')['rating'].mean())
rating.head()
```

```python
#creating no. of rating user gave
rating['number_of_ratings'] = movie_data.groupby('title')['rating'].count()
rating.head()
```

```python
#Now plotting the JointPlot
sns.jointplot(x='rating',y='number_of_ratings',data=rating)
```

```python
# Creating user item interaction matrix
movie_matrix = movie_data.pivot_table(index='userId',columns='title',values='rating')
```

```python
movie_matrix.head()
```

```python
#most rated movies so sorting the rating values
rating.sort_values('number_of_ratings', ascending=False).head(10)
```

# Making Recommendation for sample movie Ex: Solaris

```python
#now fetching ratings for solaris separately
solaris_rating = movie_matrix['Solaris']
```

```python
solaris_rating
```

```python
#Now Finding correlation with similar movies
similar_to_solaris= movie_matrix.corrwith(solaris_rating)
similar_to_solaris.head(10)
```

```python
# Creating a threshold for min no of ratings
#creating a dataframe for correlation to bring in with no. of ratings
corr_solaris = pd.DataFrame(similar_to_solaris,columns=['Correlation'])
corr_solaris.dropna(inplace=True)
corr_solaris.head()
```

```python
#now joining with ratings
corr_solaris = corr_solaris.join(rating['number_of_ratings'])
corr_solaris.head()
```

```python
#now adding filter in no. of ratings column which can give only movies which has more than rating of 30 for say

corr_solaris[corr_solaris['number_of_ratings']> 30].sort_values(by='Correlation',ascending=False).head(10)
```
