```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://www.flipkart.com/search?q=tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
url
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text,'lxml')
soup
```

```python
# creating master container
master_container = soup.find_all('div',{'class':'_2kHMtA'})
master_container
```

```python
#let us get all the names of tv
prod_name = []
prod_price = []
prod_rating = []
for i in master_container:
    name = i.find('div',{'class':'_4rR01T'}).text
    price = i.find('div',{'class':'_30jeq3 _1_WHN1'}).text
    rating = i.find('div',{'class':'_3LWZlK'})
    if rating!= None:
        prod_rating.append(rating.text)
    else:
        prod_rating.append("NA")
    prod_name.append(name)
    prod_price.append(price)
```

### Scrapping data from multiple pages

```python
prod_name = []
prod_price = []
prod_rating = []
for i in range(1,11):
    url = 'https://www.flipkart.com/search?q=tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(i)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'lxml')
    # creating master container
    master_container = soup.find_all('div',{'class':'_2kHMtA'})
    for i in master_container:
        name = i.find('div',{'class':'_4rR01T'}).text
        price = i.find('div',{'class':'_30jeq3 _1_WHN1'}).text
        rating = i.find('div',{'class':'_3LWZlK'})
        if rating!= None:
            prod_rating.append(rating.text)
        else:
            prod_rating.append("NA")
        prod_name.append(name)
        prod_price.append(price)
```

```python
len(prod_name),len(prod_price),len(prod_rating)
```

```python
final_df = pd.DataFrame({'Product Name':prod_name,
                         'Product Price':prod_price,
                         'Product Rating':prod_rating})
```

### Scrap data from a html table

```python
url = 'https://www.worldometers.info/world-population/'
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, 'lxml')
soup
```

```python
table = soup.find('table',{'class':'table table-striped table-bordered table-hover table-condensed table-list'})
table
```

```python
table.find_all('th')
```

```python
headers = []
for i in table.find_all('th'):
    col_name = i.text
    headers.append(col_name)
```

```python
headers
```

```python
final_df = pd.DataFrame(columns=headers)
final_df
```

```python
table.find_all('tr')[1:]
```

```python
len(final_df)
```

```python
for i in table.find_all('tr')[1:]:
    row_data = i.find_all('td')
    rows = [t.text for t in row_data]
    length = len(final_df)
    final_df.loc[length] = rows
```

```python
final_df
```

```python
final_df.to_csv('Population Data.csv',index=False)
```
